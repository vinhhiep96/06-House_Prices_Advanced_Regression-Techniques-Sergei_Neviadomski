# import library

```python
from google.colab import auth
import gspread
from google.auth import default
import pandas as pd
import numpy as np
from plotly.subplots import make_subplots
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')

auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

sh=gc.open_by_url('https://docs.google.com/spreadsheets/d/1gHFW4IzthaiFtGYRXEJlxMBmUnG0MDpd6WOcxsWOQWQ/edit?usp=drive_link')
df = pd.DataFrame(sh.get_worksheet(0).get_all_records())
```

# Airport by Country

```python
df1=df.loc[:10,:] # 10 values
df1['Airport Name 1']=df1['Airport Name'].str.replace("Airport",'')
fig=px.treemap(df1, path=['Country Name','Airport Name 1'])
fig.show()
```

#Airline Passenger by Gender and Continents

```python
df1 = df.head(500) #passenger = 500
df1.head(2)
def subplots_def(a,b):
  fig = make_subplots(cols=2,rows=1, subplot_titles=['Gender','Continents'], )
  a_df = df1[a].value_counts()
  b_df=df1[b].value_counts()
  fig.add_trace(go.Bar(x=a_df.index,y=a_df.values, name='Gender'), row=1, col=1)
  fig.add_trace(go.Bar(x=b_df.index,y=b_df.values, name='Continents'), row=1, col=2)
  fig.update_layout(title_text='Airline passenger by Gender and Continents')
  fig.show()
subplots_def('Gender','Continents')
```

#Airline Passenger by Age and Gender

```python
fig=px.histogram(df1, x='Age', color='Gender', nbins=100, title='Airline passenger by Age and Gender') # passenger = 500
fig.update_xaxes(nticks=50)
fig.show()
```

#Airline Passenger by Continents and Country

```python
df2=df.loc[:10,:] # passenger = 10
df2=df2.value_counts(['Continents', 'Country Name']).reset_index().rename(columns={0:"Count"})
fig=px.sunburst(df2, path=['Continents', 'Country Name'], values='Count', color='Count', color_continuous_scale='RdBu', title='Airline passenger by continents and country')
fig.show()
```

#2022 Flight Status in US

```python
df4=df.copy()
df4.columns=df4.columns.str.replace(' ','_')
df4=df4.query('Airport_Country_Code == "US"')
df4['count']=1

a =pd.to_datetime(np.arange(12), unit='D', origin=pd.Timestamp('2022-01-01'))
df4['Departure_Date']=pd.to_datetime(df4['Departure_Date'])

df4=df4.groupby(['Departure_Date','Flight_Status'])['count'].agg('count').reset_index()
df4=df4.loc[df4['Departure_Date'].isin(a)]
fig = px.bar(df4,"Departure_Date",'count', 'Flight_Status' )
fig.update_layout(
    xaxis = dict(
        tickmode = 'array',
        tickvals=df4["Departure_Date"].array,
    )
)

fig.show()
```

#10 Passenger by Age, Continental and National

```python
df3 = df.head(10)
df3['Continents_1'] = df3['Continents'].astype('category')
d1 = dict(enumerate(df3['Continents_1'].cat.categories))
df3['Continents_1'] = df3['Continents_1'].cat.codes

df3['Nationality_1']=df3['Nationality'].astype('category')
d2=dict(enumerate(df3['Nationality_1'].cat.categories))
df3['Nationality_1']=df3['Nationality_1'].cat.codes

d1_dict=pd.DataFrame.from_dict([d1])
d2_dict=pd.DataFrame.from_dict([d2])
print('values of Continents')
print(d1_dict)
print('-----')
print('values of Nationality:')
print(d2_dict)

fig=px.scatter_3d(df3,x='Age', y='Continents_1',z='Nationality_1', color='Nationality_1')
fig.show()
```
